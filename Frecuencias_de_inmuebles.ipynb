#Reporte de Análisis IV

```python
import pandas as pd
```

```python
#@title Importar tabla
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#@title Seleccione solamente los inmuebles clasificados con tipo 'Departamento'
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0]
print('El número de inmuebles clasificados con tipo Departamento -> {}'.
      format(n1))
```

```python
#@title Seleccione los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' o  'Casa de villa'
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
print('Cantidad de los inmuebles clasificados con tipos Casa, Casa en condominio o  Casa de villa-> {}'.
      format(n2))
```

```python
#@title Seleccione los inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites
seleccion = (datos['Area']>= 60) & (datos['Area']<= 100)
n3 = datos[seleccion].shape[0]
print('Número de los inmuebles con áreas entre 60 y 100 metros cuadrados, incluyendo los limites -> {}'.
      format(n3))
```

```python
#@title Seleccione los inmuebles que tengan por los menos 4 cuartos y alquiler menor que $2.000.00
seleccion = (datos['Cuartos']>= 4) & (datos['Valor']<= 2000)
n4 = datos[seleccion].shape[0]
print('Numero de inmuebles que tengan por los menos 4 cuartos y alquiler menor que $2.000,00 -> {}'.
      format(n4))
```
