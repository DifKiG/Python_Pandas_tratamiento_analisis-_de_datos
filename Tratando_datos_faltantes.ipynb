##Excluyendo valores Nulos

```python
import pandas as pd
datos = pd.read_csv('/content/alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#@title Metodos para indicar valores nulo o NaN
# Indica con verdadero don de hay null 
datos.isnull()
```

```python
#devuelve con false donde hay valores de null
datos.notnull()
```

```python
datos.info()
```

```python
#@title Dentro del DF aplica esta condicion y tra solo los que son null
datos[datos['Valor'].isnull()]
```

```python
#el metodo dropna elimina los valores NaN y el inplace actualiza eso en el DF
A = datos.shape[0]
datos.dropna(subset = ['Valor'], inplace = True)
B = datos.shape[0]
A - B
```

```python
datos[datos['Valor'].isnull()]
```

#**Tratamiento De Datos Condicional**

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
#La tilde '~' trae todo lo contrario a la linea seleccion, es decir trae esto (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
A = datos.shape[0]
datos = datos[~seleccion] 
B = datos.shape[0]
A - B
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos
```

```python
#@title eliminar los valores nulos con el metodo de .fillna 
#Se le indica con un diccionario las dos columnas que tienen valores nulos
datos = datos.fillna({'Mantenimiento' : 0, 'Impuesto' : 0})
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```

```python
datos.info()
```

```python
datos.to_csv('alquiler_residencial.csv', sep = ';', index = False)
```

#Metodos de Interpolacion

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
s.fillna(0)
```

```python
s
```

```python
#El metodo 'ffill' reemplaza el valor nulo con el valor anterior
s.fillna(method = 'ffill')
```

```python
s
```

```python
#El metodo 'bfill' reemplaza el valor nulo con el valor anterior empezando de abajo para arriba
s.fillna(method = 'bfill')
```

```python
# Se reemplaza los nulos con la media
s.fillna(s.mean())
```

```python
#El metodo 'ffill' reemplaza el valor nulo con el valor anterior limitando su accion solo 1 vez por sustitucion
s1 = s.fillna(method = 'ffill', limit = 1)
```

```python
s1
```

```python
s2 = s1.fillna(method = 'bfill', limit = 1)
```
