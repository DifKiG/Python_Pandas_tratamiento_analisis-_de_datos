##**Inmuebles residenciales**
### separar los residenciales de los no residenciales

```python
import pandas as pd
```

```python
datos = pd.read_csv('/content/drive/MyDrive/Python_Pandas_Tratamiento_y_análisis_de_datos/alquiler.csv', sep = ';')
```

```python
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial 
```

```python
datos.head()
```

```python
seleccion = datos['Tipo'].isin(residencial)
seleccion
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
#Hasta ahora se han eliminado los datos de los inmuebles que no son residenciales,
#se filtran los duplicados del DF por medio del metodo .drop_duplicate, y el resultado se almacena en
#la lista en una nueva varible llamada residencial
#se almacena los datos con la propiedad .isin en una nueva variable llamada seleccion
#para  que comprueba la existencia de un valor true en una lista
#almacenado los resultados de seleccion en una nueva variable llamada datos_residencial
```

```python
#para que los indices inicien en orden en el nuevo objeto llamado datos_residencial se llama .index = range(el tamaño del objeto)
datos_residencial.shape[0]
```

```python
datos_residencial.index = range(datos_residencial.shape[0])
datos_residencial
```

### **El siguiente ejercicio es un de un dataframe**
El código crea un DataFrame con una columna llamada N y le asigna los números del 0 al 10. Con el método isin(), seleccionamos sólo los números pares de N. 

```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

```python
seleccion= df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[seleccion]
df
```
