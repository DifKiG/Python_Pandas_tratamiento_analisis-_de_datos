##**Creando Nuevas Variables**


1.   Crear el valor bruto
2.   Crear el valor por m2
3.   Crear el valor bruto por m2







```python
import pandas as pd
datos = pd.read_csv('/content/drive/MyDrive/Python_Pandas_Tratamiento_y_análisis_de_datos/Datos/alquiler_residencial.csv',
            sep = ';')
datos.head(10)
```

```python
#@title Creando nueva variable 'Valor Bruto' a partir de la suma de las 3 columnas
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos.head(10)
```

```python
#@title Creando variable 'Valor m2' apartir de la división de 'Valor' y 'Area'
datos ['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head()
```

```python
#@title Redondera a dos cifras decimales
datos ['Valor m2'] = datos ['Valor m2'].round(2)
datos.head(10)
```

```python
#@title Creando variable 'Valor Bruto m2' apartir de la división de 'Valor Bruto' y 'Valor m2'
datos ['Valor Bruto m2'] = ((datos['Valor Bruto']) / (datos['Area'])).round(2)
datos.head(10)
```

```python
#@title todo lo que se llame 'Casa', 'Casa en condominio' o 'Casa en Villa' el resultado grupal es 'Casa' de lo contrario es 'Departamento'
casa = ['Casa', 'Casa en condominio', 'Casa de villa']
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
datos
```
